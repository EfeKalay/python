```python
def akrepYelkovanAci(saat,dakika):
    if(saat >= 12):
        saat = saat - 12
    saatAci = saat * (360/12)
    dakikaAci = dakika * (360/60)
    dakikaFarki =  (30 / 360) * dakika * 6
    saatAci = saatAci + dakikaFarki
    aci = saatAci - dakikaAci
    return aci

a = akrepYelkovanAci(6,30)
print(a)
```

```python
for saat in range(1,13):
    for dakika in range(0,60):
        angle = akrepYelkovanAci(saat,dakika)
        if (angle < 5 and angle>-5):
            print(saat,":",dakika," --> ",angle)
#saniye mutlaka hesaba katılmalı !!
```
